```python
# DATA CLEANING

df['Luas Lahan'] = pd.to_numeric(df['Luas Lahan'], errors='coerce')
df['Produksi'] = pd.to_numeric(df['Produksi'], errors='coerce')
```

```python
# Fungsi untuk menjumlahkan semua angka dalam satu sel
def sum_numbers(cell):
    if pd.isna(cell):  # Jika NaN
        return np.nan
    # Cari semua angka (desimal diperbolehkan)
    numbers = re.findall(r"[0-9]*\.?[0-9]+", str(cell))
    return sum(map(float, numbers)) if numbers else np.nan

# Bersihkan kolom Luas Lahan & Produktivitas
df['Luas Lahan Keseluruhan'] = df['Luas Lahan'].apply(sum_numbers)
df['Produksi Keseluruhan'] = df['Produksi'].apply(sum_numbers)
```

```python
def gini(array):
    array = np.array(array, dtype=np.float64)
    if np.amin(array) < 0:
        array -= np.amin(array)
    array += 1e-9  # untuk menghindari pembagian nol
    array = np.sort(array)
    n = len(array)
    index = np.arange(1, n + 1)
    return ((np.sum((2 * index - n - 1) * array)) / (n * np.sum(array)))
```

```python
# Hitung rasio gini untuk tiap desa
rasio_gini = (
    df.groupby('Desa/Kelurahan')['Luas Lahan Keseluruhan']
    .apply(lambda x: gini(x.dropna()))
    .reset_index(name='Rasio Gini')
)
```

```python
# Fungsi scoring berdasarkan aturan di gambar
def skor_gini(g):
    if g > 0.5:
        return 0
    elif g > 0.4:
        return 1
    else:
        return 4

rasio_gini['Skor'] = rasio_gini['Rasio Gini'].apply(skor_gini)
```

```python
rasio_gini
```

```python
IKP = rasio_gini[['Desa/Kelurahan']].copy()
IKP['I.1.1.1'] = rasio_gini['Skor']
```

```python
sumber_benih = df[['ID']].copy()
sumber_benih['Desa/Kelurahan'] = df['Desa/Kelurahan']
sumber_benih['Sumber Benih'] = df['Sumber Benih']
```

```python
sumber_benih.isna().sum()
```

```python
sumber_benih['Sumber Benih'].value_counts()
```

```python
def penggunaan_benih(pola):
    pola = str(pola).lower()
    
    kelompok_1 = ["tangkar sendiri"]
    kelompok_2 = ["beli", "bantuan pemerintah", "tidak ada"]
    
    ada_kelompok_1 = any(kata in pola for kata in kelompok_1)
    ada_kelompok_2 = any(kata in pola for kata in kelompok_2)
    
    if ada_kelompok_1 and ada_kelompok_2:
        return 0.5
    elif ada_kelompok_1:
        return 1
    elif ada_kelompok_2:
        return 0
    else:
        return None

sumber_benih['kategori'] = sumber_benih['Sumber Benih'].apply(penggunaan_benih)
```

```python
sumber_benih
```

```python
persentase_benih = (
    sumber_benih.groupby('Desa/Kelurahan')['kategori']
      .apply(lambda x: x.mean() * 100)
      .reset_index(name='Persentase Sumber Benih')
)

def skor_benih(p):
    if p == 0:
        return 0
    elif p <= 25:
        return 1
    elif p <= 50:
        return 2
    elif p <= 75:
        return 3
    else:
        return 4

persentase_benih['Skor'] = persentase_benih['Persentase Sumber Benih'].apply(skor_benih)
```
