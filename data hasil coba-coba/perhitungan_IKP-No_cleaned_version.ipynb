```python
df['Luas Lahan'] = pd.to_numeric(df['Luas Lahan'], errors='coerce')
df['Produktivitas'] = pd.to_numeric(df['Produktivitas'], errors='coerce')
```

```python
import re
import numpy as np

# Fungsi untuk menjumlahkan semua angka dalam satu sel
def sum_numbers(cell):
    if pd.isna(cell):  # Jika NaN
        return np.nan
    # Cari semua angka (desimal diperbolehkan)
    numbers = re.findall(r"[0-9]*\.?[0-9]+", str(cell))
    return sum(map(float, numbers)) if numbers else np.nan

# Bersihkan kolom Luas Lahan & Produktivitas
data_pertanian['Luas Lahan Clean'] = data_pertanian['Luas Lahan'].apply(sum_numbers)
data_pertanian['Produktivitas Clean'] = data_pertanian['Produktivitas'].apply(sum_numbers)
```

```python
# Hapus baris dengan produktivitas_clean = 0
data_pertanian = data_pertanian[data_pertanian['Produktivitas Clean'] != 0]
```

```python
data_pertanian.describe(include="all").T
```

```python
data_pertanian['Desa/Kelurahan'].value_counts()
```

```python
data_pertanian.isna().sum()
```

```python
data_pertanian['Penghasilan Kotor'] = data_pertanian['Produktivitas Clean'] * 10000 * 1000    # 10.000 per kg, dengan 1 ton = 1.000kg
```

```python
data_pertanian.describe(include="all").T
```

```python
data_pertanian['Pestisida/Herbisida/Fungisida.2'] = df['Pestisida/Herbisida/Fungisida.2']
data_pertanian['Benih.2'] = df['Benih.2']
data_pertanian['Upah Buruh.2'] = df['Upah Buruh.2']
data_pertanian['Sewa Alat dan Mesin Pertanian.2'] = df['Sewa Alat dan Mesin Pertanian.2']

data_pertanian['Ongkos Produksi'] = data_pertanian['Pestisida/Herbisida/Fungisida.2'] + data_pertanian['Benih.2'] + data_pertanian['Upah Buruh.2'] + data_pertanian['Sewa Alat dan Mesin Pertanian.2']
```

```python
data_pertanian['Penghasilan Bersih'] = data_pertanian['Penghasilan Kotor'] - data_pertanian['Ongkos Produksi']
```

```python
data_pertanian['Penghasilan per Bulan Bertani'] = data_pertanian['Penghasilan Bersih'] / 4
```

Garis Kemiskinan (Rupiah/Kapita/Bulan) Menurut Provinsi dan Daerah 

di Nusa Tenggara Timur pada Semester 2 (September) adalah: **495.180,00**

```python
garis_kemiskinan = data_pertanian[['ID']].copy()
garis_kemiskinan['Desa/Kelurahan'] = data_pertanian['Desa/Kelurahan']
garis_kemiskinan['Penghasilan per Bulan Bertani'] = data_pertanian['Penghasilan per Bulan Bertani']
```

```python
garis_kemiskinan
```

```python
garis_kemiskinan_rp = 495180

# Hitung persentase per desa
persentase = (garis_kemiskinan
              .groupby('Desa/Kelurahan')['Penghasilan per Bulan Bertani']
              .apply(lambda x: (x > garis_kemiskinan_rp).mean() * 100)
              .reset_index(name='Persentase di Atas GK'))

# Tambahkan kolom skor
def hitung_skor(p):
    if p == 0:
        return 0
    elif p <= 25:
        return 1
    elif p <= 50:
        return 2
    elif p <= 75:
        return 3
    else:
        return 4

persentase['Skor'] = persentase['Persentase di Atas GK'].apply(hitung_skor)
```

```python
persentase
```

```python
pilar1 = persentase[['Desa/Kelurahan']].copy()
pilar1['I.2.1.1'] = persentase['Skor']
```

### I.3.1. 1 Ketersediaan air untuk budidaya

```python
ketersediaan_air = df[['ID']].copy()
ketersediaan_air['Desa/Kelurahan'] = df['Desa/Kelurahan']
ketersediaan_air['Pola Tanam'] = df['Pola Tanam']
```

```python
ketersediaan_air
```

```python
ketersediaan_air.isna().sum()
```

```python
ketersediaan_air = ketersediaan_air.dropna()
```

```python
ketersediaan_air.isna().sum()
```

```python
ketersediaan_air['Pola Tanam'].value_counts()
```

```python
def skor_ketersediaan_air(pola):
    pola = str(pola).lower()
    if "tidak ada" in pola or "1 kali setahun" in pola:
        return 2
    elif "2 kali setahun" in pola or "3 kali setahun" in pola:
        return 1
    else:
        return 0  # jika tidak memenuhi kondisi di atas

ketersediaan_air['kategori'] = ketersediaan_air['Pola Tanam'].apply(skor_ketersediaan_air)
```

```python
ketersediaan_air
```

```python
persentase_air = (
    ketersediaan_air.groupby('Desa/Kelurahan')['kategori']
      .apply(lambda x: (x == 1).mean() * 100)
      .reset_index(name='Persentase Air Cukup')
)

def skor_air(p):
    if p == 0:
        return 0
    elif p <= 25:
        return 1
    elif p <= 50:
        return 2
    elif p <= 75:
        return 3
    else:
        return 4

persentase_air['Skor'] = persentase_air['Persentase Air Cukup'].apply(skor_air)
```

```python
persentase_air
```

```python
pilar1['I.3.1.1'] = persentase_air['Skor']
```
