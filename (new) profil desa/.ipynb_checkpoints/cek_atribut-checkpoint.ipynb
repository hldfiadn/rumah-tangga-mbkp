```python
# Mengecek kolom yang memiliki duplikasi kata (kesahalan input)

def ada_duplikasi_kata(teks):
    if pd.isna(teks):
        return False
    kata_kata = teks.split()
    return len(kata_kata) != len(set(kata_kata))

# Daftar kolom yang mau dikecualikan
skip_cols = ['Luas Lahan', 'Produktivitas']

# Cek semua kolom kecuali yang di-skip
for col in df.columns:
    if col in skip_cols:
        continue
    if df[col].astype(str).apply(ada_duplikasi_kata).any():
        print(f"Kolom '{col}' punya sel dengan duplikasi kata.")
```

```python
df.to_csv("df_sebelum_hapus_duplikasi_kata.csv", index=False)
```

```python
def hapus_duplikasi_kata(teks):
    if pd.isna(teks): 
        return teks  # biarkan NaN tetap NaN
    kata_kata = teks.split()
    kata_unik = list(dict.fromkeys(kata_kata))  # hapus duplikat sambil jaga urutan
    return ' '.join(kata_unik)

# Hapus duplikasi kata hanya untuk kolom yang tidak di-skip
for col in df.columns:
    if col in skip_cols:
        continue
    if df[col].astype(str).apply(ada_duplikasi_kata).any():
        df[col] = df[col].astype(str).apply(hapus_duplikasi_kata)

print("Semua kolom (kecuali yang dikecualikan) telah dibersihkan dari duplikasi kata.")
```

# Numerin & Categoric

```python
# daftar kolom numerik
num_cols = [
    "Pendapatan Bulanan (Rp.)", "Iuran BPJS", "Obat", "Pulsa Ponsel", 
    "Luas Lahan", "Produktivitas", "Pengairan", "Pupuk",
    "Pestisida/Herbisida/Fungisida", "Benih", "Upah Buruh",
    "Sewa Alat dan Mesin Pertanian", "Sapi", "Kerbau", "Kambing",
    "Ayam.3", "Bebek", "Babi", "Kuda", "Jumlah Hasil (Kg)", 
    "Jumlah Hasil", "Luas Lahan Usaha m2", "Luas Lahan Rumah m2",
    "Luas Bangunan Rumah m2", "Jumlah Lansia"
]

# konversi pakai pd.to_numeric biar aman
for col in num_cols:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ".").str.replace(r"[^0-9\.-]", "", regex=True), errors="coerce")

# cek hasil
print(df[num_cols].dtypes)
```

```python
import numpy as np
import pandas as pd
import re

_num_pattern = re.compile(r'^\s*[+-]?\d+(?:[.,]\d+)?\s*$')  # dukung koma/titik desimal

def sum_semicolon_numbers(cell):
    # NaN -> NaN
    if pd.isna(cell):
        return np.nan

    s = str(cell).strip()
    if s == '':
        return np.nan  # sel benar-benar kosong

    # Normalisasi delimiter & desimal
    s = s.replace('；', ';')           # fullwidth semicolon -> normal
    parts = [p.strip() for p in s.split(';')]

    total = 0.0
    found_number = False
    for p in parts:
        if p == '':         # kosong di antara ; -> 0 (skip saja)
            continue
        p2 = p.replace(',', '.')  # dukung "0,5"
        if _num_pattern.match(p2):
            total += float(p2)
            found_number = True
        else:
            # token tidak numerik -> abaikan (anggap 0)
            # kalau mau anggap invalid -> set found_number=False dan break lalu return np.nan
            pass

    return total if found_number else np.nan
```

```python
for col in num_cols:
    df[col] = df[col].apply(sum_semicolon_numbers)

# pastikan hasilnya numeric
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
```

## Menangani missing value

```python
df[num_cols] = df[num_cols].fillna(0)
```

catatan :

untuk penanganan mmsising value "numerik" bisa diatasi dengan imputasi nilai sentral (median, modus), spesifik, ataupun advance. tetapi hal ini harus didasari **background knowledge** yang benar.

dalam kasus ini, tergantung hasil akhir apa yang diinginkan? apakah persenan orang yang memiliki kuda, misalnya. kalau seperti itu, maka imputasi dengan nilai 0 (fillna) adalah yang terbaik.

untuk sementara, missing value data numerik kita ubah menjadi 0. 


> Menggunakan 0 sebagai imputasi bisa menimbulkan bias atau distorsi jika:

>Nilai 0 Bukan Representasi Realitas: Jika nilai yang hilang seharusnya adalah nilai positif/negatif yang signifikan (misalnya rata-rata gaji), menggantinya dengan 0 akan menarik rata-rata kolom ke bawah dan mengubah distribusi data secara drastis, menyebabkan model prediksi Anda menjadi bias.
