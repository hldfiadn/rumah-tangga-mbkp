```python
import numpy as np
import pandas as pd
import re

_num_pattern = re.compile(r'^\s*[+-]?\d+(?:[.,]\d+)?\s*$')  # dukung koma/titik desimal

def sum_semicolon_numbers(cell):
    # NaN -> NaN
    if pd.isna(cell):
        return np.nan

    s = str(cell).strip()
    if s == '':
        return np.nan  # sel benar-benar kosong

    # Normalisasi delimiter & desimal
    s = s.replace('；', ';')           # fullwidth semicolon -> normal
    parts = [p.strip() for p in s.split(';')]

    total = 0.0
    found_number = False
    for p in parts:
        if p == '':         # kosong di antara ; -> 0 (skip saja)
            continue
        p2 = p.replace(',', '.')  # dukung "0,5"
        if _num_pattern.match(p2):
            total += float(p2)
            found_number = True
        else:
            # token tidak numerik -> abaikan (anggap 0)
            # kalau mau anggap invalid -> set found_number=False dan break lalu return np.nan
            pass

    return total if found_number else np.nan
```

```python
df['Luas Lahan Keseluruhan'] = df['Luas Lahan'].apply(sum_semicolon_numbers)
df['Produksi Keseluruhan']   = df['Produksi'].apply(sum_semicolon_numbers)
```

```python
def gini(array):
    array = np.array(array, dtype=np.float64)
    if np.amin(array) < 0:
        array -= np.amin(array)
    array += 1e-9  # untuk menghindari pembagian nol
    array = np.sort(array)
    n = len(array)
    index = np.arange(1, n + 1)
    return ((np.sum((2 * index - n - 1) * array)) / (n * np.sum(array)))
```

```python
# Hitung rasio gini untuk tiap desa
rasio_gini = (
    df.groupby('Desa/Kelurahan')['Luas Lahan Keseluruhan']
    .apply(lambda x: gini(x.dropna()))
    .reset_index(name='Rasio Gini')
)
```

```python
# Fungsi scoring berdasarkan aturan di gambar
def skor_gini(g):
    if g > 0.5:
        return 0
    elif g > 0.4:
        return 1
    else:
        return 4

rasio_gini['Skor'] = rasio_gini['Rasio Gini'].apply(skor_gini)
```

```python
# PERHITUNGAN IKP

def sum_semicolon_numbers(cell):
    if pd.isna(cell):
        return 0.0  # kosong dianggap 0
    parts = str(cell).split(';')
    total = 0.0
    for p in parts:
        p = p.strip()
        if p == '':
            continue
        try:
            total += float(p.replace(',', '.'))
        except:
            pass
    return total

# Kolom-kolom biaya
cols_biaya = [
    'Pupuk',
    'Pestisida/Herbisida/Fungisida',
    'Benih',
    'Upah Buruh',
    'Sewa Alat dan Mesin Pertanian'
]

# Konversi semua kolom biaya ke total float
for col in cols_biaya:
    df[col] = df[col].apply(sum_semicolon_numbers)

# Hitung Ongkos Produksi
df['Ongkos Produksi'] = df[cols_biaya].sum(axis=1)
```

```python
def posisi_pertanian(s):
    # rapikan delimiter ; dan spasi
    s = str(s).strip()
    s = re.sub(r'\s*;\s*', ';', s)
    s = re.sub(r';+$', '', s)

    # pecah jadi list
    parts = [p.strip() for p in s.split(';') if p.strip()]

    # cari posisi target
    try:
        idx = parts.index("Pertanian dan Perkebunan")
        return idx + 1   # posisi 1-based
    except ValueError:
        return 0         # kalau tidak ada target

# contoh pemakaian
data_limbah['posisi'] = data_limbah['Sumber Limbah'].apply(posisi_pertanian)
```

```python
import numpy as np

def ambil_elemen_ke(s, pos):
    parts = [p.strip() for p in str(s).split(';') if p.strip()]
    if pos > 0 and pos <= len(parts):
        return parts[pos-1]   # 1-based index
    return np.nan

# buat kolom posisi
data_limbah['posisi'] = data_limbah['Sumber Limbah'].apply(posisi_pertanian)

# ambil hanya elemen sesuai posisi
data_limbah['Sumber Limbah (posisi)'] = data_limbah.apply(
    lambda row: ambil_elemen_ke(row['Sumber Limbah'], row['posisi']), axis=1
)

data_limbah['Pengolahan (posisi)'] = data_limbah.apply(
    lambda row: ambil_elemen_ke(row['Pengolahan'], row['posisi']), axis=1
)
```

```python
NTP = pd.concat([df[['ID', 'Desa/Kelurahan', 'Konsumsi Bulanan (Beli)', 'Penghasilan per Bulan Bertani']]])
```

```python
# DATA CLEANING

def sum_konsumsi_cell(s):
    if pd.isna(s):
        return 0.0
    total = 0.0
    for part in str(s).split(';'):
        t = part.strip()
        if not t:
            continue
        # ambil hanya angka, koma, titik, minus
        t = re.sub(r'[^\d,.\-]', '', t)
        # ubah format: hilangkan pemisah ribuan, koma jadi desimal
        t = t.replace('.', '').replace(',', '.')
        try:
            total += float(t)
        except ValueError:
            # kalau bukan angka valid, anggap 0
            continue
    return total

# hasil total per baris
NTP['Konsumsi Bulanan'] = NTP['Konsumsi Bulanan (Beli)'].apply(sum_konsumsi_cell)
```

```python
def ringkas_kalori_per_kelompok(df_kal: pd.DataFrame) -> pd.DataFrame:
    id_cols = ["ID", "Desa/Kelurahan"]
    out = df_kal[id_cols].copy()

    # total per kelompok (hanya kolom yang ada)
    for nama_kelompok, cols in kelompok_cols.items():
        cols_ada = [c for c in cols if c in df_kal.columns]
        if cols_ada:
            out[f"total_kalori_{nama_kelompok}"] = df_kal[cols_ada].fillna(0).sum(axis=1)
        else:
            out[f"total_kalori_{nama_kelompok}"] = 0.0

    # Total_Kalori = jumlah dari 9 kelompok (BUKAN semua 'kalori_')
    kol_kelompok = [f"total_kalori_{k}" for k in kelompok_cols.keys()]
    out["Total_Kalori"] = out[kol_kelompok].sum(axis=1, min_count=1)

    # proporsi per kelompok (% dari total)
    for nama_kelompok in kelompok_cols.keys():
        tot_k = out[f"total_kalori_{nama_kelompok}"]
        out[f"persen_{nama_kelompok}"] = np.where(
            out["Total_Kalori"] > 0, (tot_k / out["Total_Kalori"]) * 100, 0.0
        )

    # rapikan kolom
    kol_kelompok_sorted = sorted(kol_kelompok)
    kol_persen   = sorted([f"persen_{k}" for k in kelompok_cols.keys()])
    out = out[["ID", "Desa/Kelurahan", "Total_Kalori"] + kol_kelompok_sorted + kol_persen].round(2)
    return out
```

```python
# daftar kategori mengikuti keys di kelompok_cols
kategori = list(kelompok_cols.keys())

# proporsi IDEAL (%)
ideal = {
    "padi-padian":             50,
    "umbi-umbian":              6,
    "pangan hewani":           12,
    "minyak dan lemak":        10,
    "buah dan biji berminyak":  3,
    "kacang-kacangan":          5,
    "gula":                     5,
    "sayur dan buah":           6,
    "lainnya (bumbu)":          3,
}

# helper: nama kolom persen & total-kalori di ringkasan
persen_col = {k: f"persen_{k}" for k in kategori}
total_col  = {k: f"total_kalori_{k}" for k in kategori}
```

```python
def skor_pph(x: float) -> int:
    if x < 40:   return 0
    if x <= 60:  return 1
    if x <= 75:  return 2
    if x <= 90:  return 3
    return 4
```

```python
def pph_desa_from_totals(ringkasan_kelompok: pd.DataFrame) -> pd.DataFrame:
    # kolom total_kalori_* untuk 9 kategori
    kol_tot = [f"total_kalori_{k}" for k in kategori]
    # pastikan ada (kalau belum, jalankan ringkas_kalori_per_kelompok dulu)
    missing = [c for c in kol_tot if c not in ringkasan_kelompok.columns]
    if missing:
        raise ValueError(f"Kolom hilang di ringkasan: {missing}")

    # 1) agregasi total energi per-kelompok ke level desa
    agg = (ringkasan_kelompok
           .groupby("Desa/Kelurahan", as_index=False)[kol_tot]
           .sum())

    # 2) total energi desa (hanya dari 9 kelompok)
    agg["Total_Kalori_Desa"] = agg[kol_tot].sum(axis=1, min_count=1)

    # 3) persen per-kelompok di desa
    for k in kategori:
        agg[f"persen_{k}"] = np.where(
            agg["Total_Kalori_Desa"] > 0,
            (agg[f"total_kalori_{k}"] / agg["Total_Kalori_Desa"]) * 100.0,
            0.0
        )

    # 4) deviasi & PPH desa
    for k in kategori:
        agg[f"deviasi_{k}"] = (agg[f"persen_{k}"] - ideal[k]).abs()

    dev_cols = [f"deviasi_{k}" for k in kategori]
    persen_cols = [f"persen_{k}" for k in kategori]

    agg["total_deviasi_pph"] = agg[dev_cols].sum(axis=1)
    agg["PPH_Desa"] = (100 - agg["total_deviasi_pph"]).clip(0, 100)
    agg["Skor_PPH_Desa"] = agg["PPH_Desa"].apply(skor_pph).astype("int32")
    agg["cek_jumlah_persen"] = agg[persen_cols].sum(axis=1)

    # 5) keluaran ringkas per desa (kalau mau tampilkan persen per kelompok, tambahkan persen_cols)
    out_cols = ["Desa/Kelurahan", "PPH_Desa", "Skor_PPH_Desa", "cek_jumlah_persen"]
    return agg[out_cols].round(2)
```

### III.1.1.2 Indeks Shanon Weiner / Indeks Keanekaragaman

```python
carb_cols_raw = [
    "kalori_200g_Biskuit_Roti_Kue_Kemasan",
    "kalori_kg_Beras_Lokal_Sawah_Ladang",
    "kalori_kg_Beras_Toko",
    "kalori_kg_Jagung_Hibrida",
    "kalori_kg_Jagung_Lokal",
    "kalori_kg_Jewawut_Weteng",
    "kalori_kg_Singkong",
    "kalori_kg_Umbi-umbian_lain",
    "kalori_bungkus_Mie_Instan",
    "kalori_kg_Pisang",
]

protein_cols_raw = [
    "kalori_kg_Ayam",
    "kalori_kg_Daging_Sapi_Kerbau_Kambing_Babi,dll",
    "kalori_kg_Ikan_dan_Boga_Laut_Kering",
    "kalori_kg_Ikan_dan_Boga_Laut_Segar",
    "kalori_kg_Susu_Bubuk_UHT",
    "kalori_kg_Telor",
    "kalori_kg_Kacang-kacangan",
    "kalori_kg_Tahu_dan_Tempe",
]
```

```python
def _H_from_row(vals: np.ndarray) -> float:
    v = np.asarray(vals, dtype=float)
    v = np.nan_to_num(v, nan=0.0, posinf=0.0, neginf=0.0)
    N = v.sum()
    if N <= 0:
        return 0.0
    p = v / N
    p = p[p > 0]
    return float(-np.sum(p * np.log(p)))   # natural log

def _skor_shannon(h: float) -> int:
    if h < 2:      return 0
    elif h <= 2.5: return 1
    elif h <= 3.0: return 2
    elif h <= 3.5: return 3
    else:          return 4

def shannon_weiner_per_desa(df_kal: pd.DataFrame,
                            carb_cols=carb_cols_raw,
                            protein_cols=protein_cols_raw) -> pd.DataFrame:
    # gunakan hanya kolom yang ada di df_kal
    carb_cols    = sorted(set([c for c in carb_cols if c in df_kal.columns]))
    protein_cols = sorted(set([c for c in protein_cols if c in df_kal.columns]))
    all_cols     = sorted(set(carb_cols + protein_cols))
    if not all_cols:
        raise ValueError("Tidak ada kolom komoditas yang ditemukan di df_kal.")

    # siapkan & pastikan numerik
    tmp = df_kal[["Desa/Kelurahan"] + all_cols].copy()
    tmp[all_cols] = tmp[all_cols].apply(pd.to_numeric, errors="coerce").fillna(0.0)

    # agregasi total kkal per komoditas di level desa
    agg = tmp.groupby("Desa/Kelurahan", as_index=False).sum()

    # H' gabungan (karbo + protein)
    agg["H_all"] = [_H_from_row(row) for row in agg[all_cols].to_numpy()]
    agg["S_all"] = (agg[all_cols] > 0).sum(axis=1)
    agg["H_max"] = np.where(agg["S_all"] > 0, np.log(agg["S_all"]), 0.0)
    agg["Evenness"] = np.where(agg["H_max"] > 0, agg["H_all"] / agg["H_max"], 0.0)

    # opsional: H' per blok (buat analisis)
    if carb_cols:
        agg["H_karbo"] = [_H_from_row(row) for row in agg[carb_cols].to_numpy()]
        agg["S_karbo"] = (agg[carb_cols] > 0).sum(axis=1)
    else:
        agg["H_karbo"] = 0.0; agg["S_karbo"] = 0

    if protein_cols:
        agg["H_protein"] = [_H_from_row(row) for row in agg[protein_cols].to_numpy()]
        agg["S_protein"] = (agg[protein_cols] > 0).sum(axis=1)
    else:
        agg["H_protein"] = 0.0; agg["S_protein"] = 0

    # skor 0–4 berdasarkan H_all
    agg["Skor_Shannon"] = agg["H_all"].apply(_skor_shannon).astype("int32")

    return agg[[
        "Desa/Kelurahan",
        "H_all", "H_max", "Evenness", "Skor_Shannon", "S_all",
        "H_karbo", "S_karbo", "H_protein", "S_protein"
    ]].round(6)
```

```python
indeks_shannon_per_desa = shannon_weiner_per_desa(df_kal)
```

```python
p_num = pd.read_csv("produksi_6_kategori.csv")
kb_num = pd.read_csv("konsumsi_beli_6_kategori.csv")
ktb_num = pd.read_csv("konsumsi_tidak_beli_6_kategori.csv")
```

```python
# --- 4) util untuk cek "ada nilai > 0 di salah satu kolom kandidat" ---
KEYS = ["ID", "Desa/Kelurahan"]

def any_pos(df, cols, base_index):
    cols = [c for c in cols if c in df.columns]
    if not cols:
        return pd.Series(False, index=base_index)
    return df[cols].fillna(0).gt(0).any(axis=1).reindex(base_index, fill_value=False)
```

```python
# --- 5) hitung skor ---
p   = p_num.set_index(KEYS)
kb  = kb_num.set_index(KEYS)
ktb = ktb_num.set_index(KEYS)
base_index = p.index.union(kb.index).union(ktb.index)

scores = {}
```

```python
# --- 2) daftar kolom yang akan dipakai per komoditas (pakai kandidat biar tahan variasi nama) ---
P_MAP = {
    "beras":       ["Beras_Lokal_(Sawah/Ladang_dalam_Kg)"],
    "jagung":      ["Jagung_Lokal", "Jagung_Hibrida"],
    "singkong":    ["Singkong"],
    "ikan_segar":  ["Ikan_dan_Boga_Laut_Segar"],
    "telor":       ["Telor"],
    "daging":      ["Daging_(Sapi/Kerbau/Kambing/Babi,dll)"],
}
KB_MAP = {
    "beras":       ["Beras_Lokal_(Sawah/Ladang_dalam_Rp)", "Beras_Toko"],
    "jagung":      ["Jagung_Lokal.2", "Jagung_Hibrida.2"],
    "singkong":    ["Singkong.2"],
    "ikan_segar":  ["Ikan_dan_Boga_Laut_Segar.2"],
    "telor":       ["Telor.2"],
    "daging":      ["Daging_(Sapi/Kerbau/Kambing/Babi,dll).2",
                    "Daging_(Sapi/Kerbau/Kambing/Babi,dll).1",
                    "Daging_(Sapi/Kerbau/Kambing/Babi,dll)"],
}
KTB_MAP = {
    "beras":       ["Beras_Lokal_(Sawah/Ladang)"],
    "jagung":      ["Jagung_Lokal.1", "Jagung_Hibrida.1"],
    "singkong":    ["Singkong.1"],
    "ikan_segar":  ["Ikan_dan_Boga_Laut_Segar.1"],
    "telor":       ["Telor.1"],
    "daging":      ["Daging_(Sapi/Kerbau/Kambing/Babi,dll).1",
                    "Daging_(Sapi/Kerbau/Kambing/Babi,dll)"],
}
```

```python
# BERAS: 100 kalau produksi / tidak beli / beli "Beras Lokal (Rp)"; 60 kalau hanya beli Beras Toko
beras_100 = (any_pos(ktb, KTB_MAP["beras"], base_index) |
             any_pos(p,   P_MAP["beras"],   base_index) |
             any_pos(kb,  ["Beras_Lokal_(Sawah/Ladang_dalam_Rp)"], base_index))
beras_60  = (~beras_100) & any_pos(kb, ["Beras_Toko"], base_index)
scores["beras"] = np.select([beras_100, beras_60], [100, 60], default=np.nan)

# JAGUNG
jagung_100 = any_pos(ktb, KTB_MAP["jagung"], base_index) | any_pos(p, P_MAP["jagung"], base_index)
jagung_60  = (~jagung_100) & any_pos(kb, KB_MAP["jagung"], base_index)
scores["jagung"] = np.select([jagung_100, jagung_60], [100, 60], default=np.nan)

# SINGKONG
singkong_100 = any_pos(ktb, KTB_MAP["singkong"], base_index) | any_pos(p, P_MAP["singkong"], base_index)
singkong_60  = (~singkong_100) & any_pos(kb, KB_MAP["singkong"], base_index)
scores["singkong"] = np.select([singkong_100, singkong_60], [100, 60], default=np.nan)

# IKAN SEGAR
ikan_100 = any_pos(ktb, KTB_MAP["ikan_segar"], base_index) | any_pos(p, P_MAP["ikan_segar"], base_index)
ikan_60  = (~ikan_100) & any_pos(kb, KB_MAP["ikan_segar"], base_index)
scores["ikan_segar"] = np.select([ikan_100, ikan_60], [100, 60], default=np.nan)

# TELOR
telor_100 = any_pos(ktb, KTB_MAP["telor"], base_index) | any_pos(p, P_MAP["telor"], base_index)
telor_60  = (~telor_100) & any_pos(kb, KB_MAP["telor"], base_index)
scores["telor"] = np.select([telor_100, telor_60], [100, 60], default=np.nan)

# DAGING
daging_100 = any_pos(ktb, KTB_MAP["daging"], base_index) | any_pos(p, P_MAP["daging"], base_index)
daging_60  = (~daging_100) & any_pos(kb, KB_MAP["daging"], base_index)
scores["daging"] = np.select([daging_100, daging_60], [100, 60], default=np.nan)
```

```python
# --- 6) rakit dataframe hasil ---
asal_bahan = (pd.DataFrame({
    "ID": [i[0] for i in base_index],
    "Desa/Kelurahan": [i[1] for i in base_index],
    **scores
})
.sort_values(["Desa/Kelurahan", "ID"])
.reset_index(drop=True))
```
