```python
# Mengecek kolom yang memiliki duplikasi kata (kesahalan input)

def ada_duplikasi_kata(teks):
    if pd.isna(teks):
        return False
    kata_kata = teks.split()
    return len(kata_kata) != len(set(kata_kata))

# Cek semua kolom
for col in df_na.columns:
    if df_na[col].astype(str).apply(ada_duplikasi_kata).any():
        print(f"Kolom '{col}' punya sel dengan duplikasi kata.")
```

```python
def hapus_duplikasi_kata(teks):
    if pd.isna(teks): 
        return teks  # biarkan NaN tetap NaN
    kata_kata = teks.split()
    kata_unik = list(dict.fromkeys(kata_kata))  # hapus duplikat sambil jaga urutan
    return ' '.join(kata_unik)

# Loop hanya kolom yang punya duplikasi kata
for col in df_na.columns:
    if df_na[col].astype(str).apply(ada_duplikasi_kata).any():
        df_na[col] = df_na[col].astype(str).apply(hapus_duplikasi_kata)

print(f"Semua kolom telah dibersihkan dari duplikasi kata.")
```
