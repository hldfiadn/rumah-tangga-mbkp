```python
import re
import pandas as pd
```

```python
df = pd.read_csv("data_rumah_tangga_all.csv")
```

```python
# mengubah format dengan menghilangkan tanda ;

df = df.replace(r';', '', regex=True).replace(r'^\s+$', '0', regex=True)
```

```python
# Mengecek kolom yang memiliki duplikasi kata (kesahalan input)

def ada_duplikasi_kata(teks):
    if pd.isna(teks):
        return False
    kata_kata = teks.split()
    return len(kata_kata) != len(set(kata_kata))

# Cek semua kolom
for col in df.columns:
    if df[col].astype(str).apply(ada_duplikasi_kata).any():
        print(f"Kolom '{col}' punya sel dengan duplikasi kata.")
```

```python
def hapus_duplikasi_kata(teks):
    if pd.isna(teks): 
        return teks  # biarkan NaN tetap NaN
    kata_kata = teks.split()
    kata_unik = list(dict.fromkeys(kata_kata))  # hapus duplikat sambil jaga urutan
    return ' '.join(kata_unik)

# Loop hanya kolom yang punya duplikasi kata
for col in df.columns:
    if df[col].astype(str).apply(ada_duplikasi_kata).any():
        df[col] = df[col].astype(str).apply(hapus_duplikasi_kata)

print(f"Semua kolom telah dibersihkan dari duplikasi kata.")
```

```python
# rename kolom Data Produksi
cols_to_convert_0 = df.columns[19:43]
df[cols_to_convert_0] = df[cols_to_convert_0].apply(pd.to_numeric, errors='coerce')
clean_cols_0 = [re.sub(r'\.\d+$', '', col) for col in cols_to_convert_0]
df.rename(columns=dict(zip(cols_to_convert_0, clean_cols_0)), inplace=True)
renamed_cols_0 = [col + '.0' for col in clean_cols_0]
df.rename(columns=dict(zip(clean_cols_0, renamed_cols_0)), inplace=True)

# rename kolom Data Konsumsi Tidak Beli
cols_to_convert_1 = df.columns[43:66]
df[cols_to_convert_1] = df[cols_to_convert_1].apply(pd.to_numeric, errors='coerce')
clean_cols_1 = [re.sub(r'\.\d+$', '', col) for col in cols_to_convert_1]
df.rename(columns=dict(zip(cols_to_convert_1, clean_cols_1)), inplace=True)
renamed_cols_1 = [col + '.1' for col in clean_cols_1]
df.rename(columns=dict(zip(clean_cols_1, renamed_cols_1)), inplace=True)

# rename kolom Data Konsumsi Beli
cols_to_convert_2 = df.columns[66:160]
df[cols_to_convert_2] = df[cols_to_convert_2].apply(pd.to_numeric, errors='coerce')
clean_cols_2 = [re.sub(r'\.\d+$', '', col) for col in cols_to_convert_2]
df.rename(columns=dict(zip(cols_to_convert_2, clean_cols_2)), inplace=True)
renamed_cols_2 = [col + '.2' for col in clean_cols_2]
df.rename(columns=dict(zip(clean_cols_2, renamed_cols_2)), inplace=True)

df.iloc[:, 19:159] = df.iloc[:, 19:159].fillna(0)
```
