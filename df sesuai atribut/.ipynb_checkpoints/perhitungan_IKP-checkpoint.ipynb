```python
import numpy as np
import pandas as pd
import re

_num_pattern = re.compile(r'^\s*[+-]?\d+(?:[.,]\d+)?\s*$')  # dukung koma/titik desimal

def sum_semicolon_numbers(cell):
    # NaN -> NaN
    if pd.isna(cell):
        return np.nan

    s = str(cell).strip()
    if s == '':
        return np.nan  # sel benar-benar kosong

    # Normalisasi delimiter & desimal
    s = s.replace('；', ';')           # fullwidth semicolon -> normal
    parts = [p.strip() for p in s.split(';')]

    total = 0.0
    found_number = False
    for p in parts:
        if p == '':         # kosong di antara ; -> 0 (skip saja)
            continue
        p2 = p.replace(',', '.')  # dukung "0,5"
        if _num_pattern.match(p2):
            total += float(p2)
            found_number = True
        else:
            # token tidak numerik -> abaikan (anggap 0)
            # kalau mau anggap invalid -> set found_number=False dan break lalu return np.nan
            pass

    return total if found_number else np.nan
```

```python
df['Luas Lahan Keseluruhan'] = df['Luas Lahan'].apply(sum_semicolon_numbers)
df['Produksi Keseluruhan']   = df['Produksi'].apply(sum_semicolon_numbers)
```

```python
# Hapus baris dengan Luas Lahan Keseluruhan = 0
df = df[df['Luas Lahan Keseluruhan'] != 0]
```

```python
df
```

```python
def gini(array):
    array = np.array(array, dtype=np.float64)
    if np.amin(array) < 0:
        array -= np.amin(array)
    array += 1e-9  # untuk menghindari pembagian nol
    array = np.sort(array)
    n = len(array)
    index = np.arange(1, n + 1)
    return ((np.sum((2 * index - n - 1) * array)) / (n * np.sum(array)))
```

```python
# Hitung rasio gini untuk tiap desa
rasio_gini = (
    df.groupby('Desa/Kelurahan')['Luas Lahan Keseluruhan']
    .apply(lambda x: gini(x.dropna()))
    .reset_index(name='Rasio Gini')
)
```

```python
# Fungsi scoring berdasarkan aturan di gambar
def skor_gini(g):
    if g > 0.5:
        return 0
    elif g > 0.4:
        return 1
    else:
        return 4

rasio_gini['Skor'] = rasio_gini['Rasio Gini'].apply(skor_gini)
```

```python
rasio_gini
```

```python
IKP = rasio_gini[['Desa/Kelurahan']].copy()
IKP['I.1.1.1'] = rasio_gini['Skor']
```

```python
# PERHITUNGAN IKP

def sum_semicolon_numbers(cell):
    if pd.isna(cell):
        return 0.0  # kosong dianggap 0
    parts = str(cell).split(';')
    total = 0.0
    for p in parts:
        p = p.strip()
        if p == '':
            continue
        try:
            total += float(p.replace(',', '.'))
        except:
            pass
    return total

# Kolom-kolom biaya
cols_biaya = [
    'Pestisida/Herbisida/Fungisida',
    'Benih',
    'Upah Buruh',
    'Sewa Alat dan Mesin Pertanian'
]

# Konversi semua kolom biaya ke total float
for col in cols_biaya:
    df[col] = df[col].apply(sum_semicolon_numbers)

# Hitung Ongkos Produksi
df['Ongkos Produksi'] = df[cols_biaya].sum(axis=1)
```

```python
sumber_benih = df[['ID']].copy()
sumber_benih['Desa/Kelurahan'] = df['Desa/Kelurahan']
sumber_benih['Sumber Benih'] = df['Sumber Benih']
```

```python
sumber_benih.isna().sum()
```

```python
sumber_benih['Sumber Benih'].value_counts()
```

```python
def penggunaan_benih(pola):
    pola = str(pola).lower()
    
    kelompok_1 = ["tangkar sendiri"]
    kelompok_2 = ["beli", "bantuan pemerintah", "tidak ada"]
    
    ada_kelompok_1 = any(kata in pola for kata in kelompok_1)
    ada_kelompok_2 = any(kata in pola for kata in kelompok_2)
    
    if ada_kelompok_1 and ada_kelompok_2:
        return 0.5
    elif ada_kelompok_1:
        return 1
    elif ada_kelompok_2:
        return 0
    else:
        return None

sumber_benih['kategori'] = sumber_benih['Sumber Benih'].apply(penggunaan_benih)
```

```python
sumber_benih
```

```python
persentase_benih = (
    sumber_benih.groupby('Desa/Kelurahan')['kategori']
      .apply(lambda x: x.mean() * 100)
      .reset_index(name='Persentase Sumber Benih')
)

def skor_benih(p):
    if p == 0:
        return 0
    elif p <= 25:
        return 1
    elif p <= 50:
        return 2
    elif p <= 75:
        return 3
    else:
        return 4

persentase_benih['Skor'] = persentase_benih['Persentase Sumber Benih'].apply(skor_benih)
```

```python
persentase_benih
```

```python
IKP['I.4.1.1'] = persentase_benih['Skor']
```
