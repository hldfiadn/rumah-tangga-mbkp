```python
import pandas as pd

# Daftar nama file sesuai urutan di gambar
file_names = [
    "1. Benihading.xlsx", "2. Riangbao.xlsx", "3. Katakeja.xlsx", "4. Waienga.xlsx", "6. TubungWalang.xlsx",
    "7. Merdeka.xlsx", "8. Dikesare.xlsx", "9. Watodiri.xlsx", "10. Hoelea I.xlsx", "11. Muruona.xlsx",
    "12. Wuakerong.xlsx", "13. Tagawiti.xlsx", "14. Lamatuka.xlsx", "15. Ile Kimok.xlsx", "16. Bour.xlsx",
    "17. Puor.xlsx", "18. Balauring.xlsx", "19. Nubamado.xlsx", "20. Labalimut.xlsx", "22. Mahal.xlsx",
    "23. Duawutun.xlsx", "24. Dulitukan.xlsx", "26. Tubuk Rajan.xlsx", "27. Pada.xlsx", "28. Paubokol.xlsx",
    "29. Kalikur.xlsx", "30. Lolong.xlsx"
]

# List untuk menampung DataFrame
df_list = []

# Loop baca semua file dengan header di baris ke-2
for file in file_names:
    df = pd.read_excel(file, header=1)  # Ambil nama kolom dari baris ke-2
    df['SourceFile'] = file
    df_list.append(df)

# Gabungkan semua DataFrame
df_combined = pd.concat(df_list, ignore_index=True)

# Tampilkan hasil awal
df_combined.head()
```

```python
df_combined = df_combined.drop(columns='SourceFile')
```

```python
# Fungsi bantu untuk mengganti spasi dengan underscore pada nama kolom
def clean_columns(df):
    df.columns = df.columns.str.replace(' ', '_')
    return df

# Ambil kolom id
id_column = df_combined.iloc[:, 1]

# Gabungkan kolom id ke masing-masing subset
Formulir_Data_Rumah_Tangga = pd.concat([id_column, clean_columns(df_combined.iloc[:, 2:14])], axis=1)
Data_Keluarga              = pd.concat([id_column, clean_columns(df_combined.iloc[:, 14:19])], axis=1)
Data_Produksi              = pd.concat([id_column, clean_columns(df_combined.iloc[:, 19:47])], axis=1)
Data_Konsumsi_Tidak_Beli   = pd.concat([id_column, clean_columns(df_combined.iloc[:, 47:74])], axis=1)
Data_Konsumsi_Beli         = pd.concat([id_column, clean_columns(df_combined.iloc[:, 74:141])], axis=1)
Data_Pertanian             = pd.concat([id_column, clean_columns(df_combined.iloc[:, 178:189])], axis=1)
Data_Limbah                = pd.concat([id_column, clean_columns(df_combined.iloc[:, 256:268])], axis=1)
Data_Aset_Keluarga         = pd.concat([id_column, clean_columns(df_combined.iloc[:, 268:274])], axis=1)

# Potong dan beri nama DataFrame sesuai index kolom
#Data_Perkebunan                = clean_columns(df_combined.iloc[:, 189:199])# 189 - 198
#Data_Peternakan                = clean_columns(df_combined.iloc[:, 199:210])# 199 - 209
#Data_Perikanan_Budidaya        = clean_columns(df_combined.iloc[:, 210:217])# 210 - 216
#Data_Perikanan_Tangkap         = clean_columns(df_combined.iloc[:, 217:260])# 217 - 259
#Pemanfaatan_Hasil_Produksi     = clean_columns(df_combined.iloc[:, 260:265])# 260 - 264
#Data_Kesehatan_Keluarga        = clean_columns(df_combined.iloc[:, 274:290])# 274 - 289

# Simpan ke file CSV sesuai nama variabel
Formulir_Data_Rumah_Tangga.to_csv("Formulir_Data_Rumah_Tangga.csv", index=False)
Data_Keluarga.to_csv("Data_Keluarga.csv", index=False)
Data_Produksi.to_csv("Data_Produksi.csv", index=False)
Data_Konsumsi_Tidak_Beli.to_csv("Data_Konsumsi_Tidak_Beli.csv", index=False)
Data_Konsumsi_Beli.to_csv("Data_Konsumsi_Beli.csv", index=False)
Data_Pertanian.to_csv("Data_Pertanian.csv", index=False)
#Data_Perkebunan.to_csv("Data_Perkebunan.csv", index=False)
#Data_Peternakan.to_csv("Data_Peternakan.csv", index=False)
#Data_Perikanan_Budidaya.to_csv("Data_Perikanan_Budidaya.csv", index=False)
#Data_Perikanan_Tangkap.to_csv("Data_Perikanan_Tangkap.csv", index=False)
#Pemanfaatan_Hasil_Produksi.to_csv("Pemanfaatan_Hasil_Produksi.csv", index=False)
Data_Limbah.to_csv("Data_Limbah.csv", index=False)
Data_Aset_Keluarga.to_csv("Data_Aset_Keluarga.csv", index=False)
#Data_Kesehatan_Keluarga.to_csv("Data_Kesehatan_Keluarga.csv", index=False)

print("✅ Semua DataFrame telah disimpan ke CSV dengan nama sesuai kategorinya.")
```

```python
import re

def hapus_duplikasi_kata(teks):
    if pd.isna(teks): return teks  # biarkan NaN tetap NaN
    # Pecah berdasarkan spasi
    kata_kata = teks.split()
    # Hilangkan duplikat yang berurutan maupun tidak
    kata_unik = list(dict.fromkeys(kata_kata))
    return ' '.join(kata_unik)

# Terapkan ke kolom Pengairan
Data_Pertanian['Pengairan'] = Data_Pertanian['Pengairan'].astype(str).apply(hapus_duplikasi_kata)
```

```python
# Strip whitespace berlebih dan ubah ke format konsisten (misal Title Case)
Data_Pertanian['Pengairan'] = Data_Pertanian['Pengairan'].str.strip().str.title()
```

```python
Data_Pertanian['Pengairan'].value_counts()
```

```python
kolom_target = ['Luas_Lahan', 'Produktivitas']

Data_Pertanian[kolom_target] = Data_Pertanian[kolom_target].apply(
    pd.to_numeric, errors='coerce'
)
```
