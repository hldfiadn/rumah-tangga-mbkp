```python
def parse_rp(x):
    """Bersihkan string rupiah (Rp, titik ribuan, koma) → float (nilai Rp)."""
    if pd.isna(x):
        return np.nan
    s = str(x)
    s = re.sub(r"[^0-9,.\-]", "", s)            # buang selain digit, koma, titik, minus
    s = re.sub(r"\.(?=\d{3}(\D|$))", "", s)     # hapus titik ribuan (contoh 12.600 → 12600)
    s = s.replace(",", ".")                     # koma → titik desimal
    try:
        return float(s)
    except:
        return np.nan

def parse_num(x):
    """Bersihkan angka umum (mis. '1,5', '3 kg', '1.200,5') → float."""
    if pd.isna(x): 
        return np.nan
    s = str(x)
    s = re.sub(r"[^0-9,.\-]", "", s)            # buang huruf/simbol selain digit . , -
    s = re.sub(r"\.(?=\d{3}(\D|$))", "", s)     # hapus titik ribuan
    s = s.replace(",", ".")                     # koma → titik
    try:
        return float(s)
    except:
        return np.nan

def make_new_name(unit, old_name):
    # contoh: 'Ayam.2' → 'Ayam_2'
    base = old_name.replace("/", "_").replace(" ", "_").replace("(", "").replace(")", "").replace(".", "_")
    base = re.sub(r"__+", "_", base).strip("_")
    return f"{unit}_{base}"
```

```python
# ------- 3) Satuan untuk masing2 kolom (apa yang mau dipakai di prefix) -------
satuan = {
    # default untuk yang per-kg
    **{k: "kg" for k, v in harga.items() if v and k not in [
        "Kelapa", "Bumbu_Kemasan", "MSG", "Lada", "Kemiri.2", "Kecap",
        "Minuman_Kemasan", "Minuman_Fermentasi", "Kopi", "Teh",
        "Mie_Instan", "Rokok_Kemasan", "Tembakau", "Biskuit/Roti/Kue_Kemasan"
    ]},
    # override
    "Kelapa": "butir",
    "Bumbu_Kemasan": "bungkus",
    "MSG": "400g",
    "Lada": "ons",
    "Kemiri.2": "ons",
    "Kecap": "botol",
    "Minuman_Kemasan": "botol",
    "Minuman_Fermentasi": "botol",
    "Kopi": "ons",
    "Teh": "bungkus",
    "Mie_Instan": "bungkus",
    "Rokok_Kemasan": "bungkus",
    "Tembakau": "ons",
    "Biskuit/Roti/Kue_Kemasan": "200g",
}
```

```python
# ------- KONVERSI: KB (semua Rp → jumlah satuan alamiah) -------
def convert_kb(df_kb):
    out = pd.DataFrame(index=df_kb.index)
    for col, h in harga.items():
        if col not in df_kb.columns:
            continue
        if h and h > 0:
            unit = satuan.get(col, "unit")
            newcol = make_new_name(unit, col)
            # jika belum ada _1/_2, paksa _2 (KB)
            if not re.search(r"_(?:1|2)$", newcol):
                newcol = newcol + "_2"
            out[newcol] = df_kb[col].apply(parse_rp) / h
    return out
```

```python
# ------- 5) KONVERSI: KTB (campuran kg & Rp) -------
# daftar kolom KTB yang SUDAH kg (dari list kamu)
ktb_sudah_kg = {
    "Beras_Lokal_(Sawah/Ladang)", "Singkong.1", "Jagung_Lokal.1", "Jagung_Hibrida.1",
    "Umbi-umbian_lain.1", "Sorgum.1", "Jewawut/Weteng.1"
}

def convert_ktb(df_ktb):
    out = pd.DataFrame(index=df_ktb.index)
    for col in df_ktb.columns:
        if col in ktb_sudah_kg:
            newcol = make_new_name("kg", col)   # akan jadi mis. 'kg_Beras_Lokal_Sawah_Ladang'
            # kalau belum ada _1/_2 → paksa _1 (KTB)
            if not re.search(r"_(?:1|2)$", newcol):
                newcol = newcol + "_1"
            out[newcol] = df_ktb[col].apply(parse_num)
            continue

        # kolom Rp di KTB → map ke harga KB (".1" → ".2")
        kb_like = col.replace(".1", ".2")
        if kb_like in harga and harga[kb_like]:
            unit = satuan.get(kb_like, "unit")
            newcol = make_new_name(unit, col)  # 'kg_Ayam_1' dll
            if not re.search(r"_(?:1|2)$", newcol):
                newcol = newcol + "_1"
            out[newcol] = df_ktb[col].apply(parse_rp) / harga[kb_like]
    return out
```

```python
# ------- 6) jalankan -------
kb_unit = convert_kb(df_kb)   # → kolom seperti kg_Tomat, bungkus_Mie_Instan, sachet_Kopi, butir_Kelapa, ons_Kemiri, ...
ktb_unit = convert_ktb(df_ktb)

# tambahkan ID & Desa di depan
kb_unit = pd.concat([df_kb[["ID", "Desa/Kelurahan"]].reset_index(drop=True),
                     kb_unit.reset_index(drop=True)], axis=1)

ktb_unit = pd.concat([df_ktb[["ID", "Desa/Kelurahan"]].reset_index(drop=True),
                      ktb_unit.reset_index(drop=True)], axis=1)
```

```python
KEYS = ["ID", "Desa/Kelurahan"]

# 1) gabungkan KB & KTB
merged = pd.merge(kb_unit, ktb_unit, on=KEYS, how="outer")

# 2) kelompokkan kolom ukuran ke "base" (hapus _1/_2, dan khusus beras lokal: hapus _dalam_Rp)
measure_cols = [c for c in merged.columns if c not in KEYS]
groups = {}
for c in measure_cols:
    base = re.sub(r"_(?:1|2)$", "", c)  # buang _1/_2
    # normalisasi khusus Beras Lokal: ..._dalam_Rp -> (hapus)
    base = base.replace("_dalam_Rp", "") if base.startswith("kg_Beras_Lokal_Sawah_Ladang") else base
    groups.setdefault(base, []).append(c)

# 3) bentuk total_units: jumlahkan pasangan (ada _1 & _2), sisanya bawa apa adanya
total_units = merged[KEYS].copy()
for base, cols in groups.items():
    has_1 = any(col.endswith("_1") for col in cols)
    has_2 = any(col.endswith("_2") for col in cols)
    if has_1 and has_2:
        total_units["konsumsi_" + base] = merged[cols].fillna(0).sum(axis=1)
    else:
        total_units["konsumsi_" + base] = merged[cols[0]]

# 4) urutan: pasangan (KTB+KB) → KTB-only → KB-only
#    (pakai logika yang sama, tapi dihitung dari kb_unit/ktb_unit)
ktb_bases = {
    re.sub(r"_(?:1|2)$", "", c).replace("_dalam_Rp", "") if re.sub(r"_(?:1|2)$", "", c).startswith("kg_Beras_Lokal_Sawah_Ladang")
    else re.sub(r"_(?:1|2)$", "", c)
    for c in ktb_unit.columns if c not in KEYS
}
kb_bases = {
    re.sub(r"_(?:1|2)$", "", c).replace("_dalam_Rp", "") if re.sub(r"_(?:1|2)$", "", c).startswith("kg_Beras_Lokal_Sawah_Ladang")
    else re.sub(r"_(?:1|2)$", "", c)
    for c in kb_unit.columns if c not in KEYS
}

both     = sorted(ktb_bases & kb_bases)
ktb_only = sorted(ktb_bases - kb_bases)
kb_only  = sorted(kb_bases  - ktb_bases)

order = []
order += [f"konsumsi_{b}" for b in both]      # pasangan dulu
order += [f"konsumsi_{b}" for b in ktb_only]  # KTB-only
order += [f"konsumsi_{b}" for b in kb_only]   # KB-only

cols_final = KEYS + [c for c in order if c in total_units.columns]
total_units = total_units[cols_final]
```

```python
# ========= 1) PETA KALORI / SATUAN =========
# KUNCI: pakai format 'satuan_Nama' (tanpa prefix 'konsumsi_')
# NILAI: kalori per 1 <satuan> persis seperti di kolom.
# - Jika kolom kg_... → nilai harus "kcal per kg"
# - Jika kolom ons_... → nilai "kcal per ons (100 g)"
# - Jika kolom 200g_/400g_ → nilai "kcal per 200 g / 400 g"
# - Jika botol/bungkus/butir/sachet → "kcal per 1 unit tsb"
kcal_per_unit = {
    # --- contoh dari daftar kamu (ANGKA CONTOH, ganti dengan angka finalmu) ---
    "kg_Timun": 80,                        # (sudah kamu sesuaikan ke per kg)
    "kg_Tomat": 240,
    "kg_Kangkung": 280,
    "kg_Sawi": 280,
    "kg_Labu": 300,
    "bungkus_Mie_Instan": 300,
    "ons_Lada": 333,                       # per ons (100 g)
    "kg_Wortel": 360,
    "kg_Cabai_Besar": 360,                 # contoh: 50 kcal/100g → 500 per kg (kamu tulis 360 di daftar contoh)
    "kg_Jeruk": 450,
    "kg_Pepaya": 460,
    "kg_Bawang_Merah": 460,
    "kg_Kubis": 510,
    "kg_Mangga": 520,
    "botol_Kecap": 540,
    "kg_Daun_Ubi": 620,
    "ons_Kemiri": 675,
    "kg_Salak": 770,
    "200g_Biskuit_Roti_Kue_Kemasan": 916,  # per 200 g
    "kg_Ikan_dan_Boga_Laut_Segar": 1000,
    "kg_Pisang": 1080,
    "kg_Bawang_Putih": 1120,
    "kg_Tahu_dan_Tempe": 1200,
    "kg_Cabai_Kecil": 1200,
    "kg_Umbi-umbian_lain": 1300,
    "kg_Daging_Sapi_Kerbau_Kambing_Babi,dll": 1500,
    "kg_Singkong": 1540,
    "kg_Telor": 1540,
    "kg_Ikan_dan_Boga_Laut_Kering": 2500,
    "kg_Ayam": 2980,
    "kg_Kental_Manis": 3430,
    "kg_Beras_Toko": 3570,
    "kg_Jewawut_Weteng": 3640,
    "kg_Jagung_Lokal": 3660,
    "kg_Jagung_Hibrida": 3660,
    "kg_Beras_Lokal_Sawah_Ladang": 3760,
    "kg_Kacang-kacangan": 5000,
    "kg_Susu_Bubuk_UHT": 5130,
    "kg_Minyak_Goreng_Kemasan_Toko": 8181,
    "kg_Minyak_Kelapa": 8700,
    "kg_Gula_Pasir": 3940,

    # --- unit lain di daftar contohmu ---
    "bungkus_Bumbu_Kemasan": 0,
    "400g_MSG": 0,
    "bungkus_Teh": 0,
    "botol_Minuman_Kemasan": 70,
    "botol_Minuman_Fermentasi": 70,
    "bungkus_Rokok_Kemasan": 100,
    "butir_Kelapa": 159,
    "kg_Bayam": 160,
    "sachet_Kopi": 241,   # isi sesuai kebutuhan
    # dst... lengkapi semua item yang muncul di total_units
}
```

```python
def konsumsi_ke_kalori(total_units: pd.DataFrame) -> pd.DataFrame:
    id_cols = ["ID", "Desa/Kelurahan"]
    kons_cols = [c for c in total_units.columns if c.startswith("konsumsi_")]

    out = total_units[id_cols].copy()

    for c in kons_cols:
        unit_name = re.sub(r"^konsumsi_", "", c)
        factor = kcal_per_unit.get(unit_name)

        if factor is None:
            # skip kalau belum ada faktor kalori
            continue

        out["kalori_" + unit_name] = total_units[c].fillna(0) * factor

    kal_cols = [k for k in out.columns if k.startswith("kalori_")]
    out["Total_Kalori"] = out[kal_cols].sum(axis=1, min_count=1)

    # urutkan & bulatkan 2 desimal
    out = out[["ID", "Desa/Kelurahan", "Total_Kalori"] + sorted(kal_cols)]
    out = out.round(2)

    return out
```

```python
KEYS = ["ID", "Desa/Kelurahan"]

# --- 1) helper untuk membersihkan angka semicolon ---
num_pat = re.compile(r'-?\d+(?:[.,]\d+)?')

def parse_sum(cell):
    if pd.isna(cell): 
        return np.nan
    nums = num_pat.findall(str(cell))
    if not nums:
        return np.nan
    return sum(float(x.replace(',', '.')) for x in nums)

def coerce_cols(df, cols):
    df2 = df.copy()
    for c in cols:
        if c in df2.columns:
            df2[c] = df2[c].apply(parse_sum)
    return df2
```

```python
# --- 2) daftar kolom yang akan dipakai per komoditas (pakai kandidat biar tahan variasi nama) ---
P_MAP = {
    "beras":       ["Beras_Lokal_(Sawah/Ladang_dalam_Kg)"],
    "jagung":      ["Jagung_Lokal", "Jagung_Hibrida"],
    "singkong":    ["Singkong"],
    "ikan_segar":  ["Ikan_dan_Boga_Laut_Segar"],
    "telor":       ["Telor"],
    "daging":      ["Daging_(Sapi/Kerbau/Kambing/Babi,dll)"],
}
KB_MAP = {
    "beras":       ["Beras_Lokal_(Sawah/Ladang_dalam_Rp)", "Beras_Toko"],
    "jagung":      ["Jagung_Lokal.2", "Jagung_Hibrida.2"],
    "singkong":    ["Singkong.2"],
    "ikan_segar":  ["Ikan_dan_Boga_Laut_Segar.2"],
    "telor":       ["Telor.2"],
    "daging":      ["Daging_(Sapi/Kerbau/Kambing/Babi,dll).2",
                    "Daging_(Sapi/Kerbau/Kambing/Babi,dll).1",
                    "Daging_(Sapi/Kerbau/Kambing/Babi,dll)"],
}
KTB_MAP = {
    "beras":       ["Beras_Lokal_(Sawah/Ladang)"],
    "jagung":      ["Jagung_Lokal.1", "Jagung_Hibrida.1"],
    "singkong":    ["Singkong.1"],
    "ikan_segar":  ["Ikan_dan_Boga_Laut_Segar.1"],
    "telor":       ["Telor.1"],
    "daging":      ["Daging_(Sapi/Kerbau/Kambing/Babi,dll).1",
                    "Daging_(Sapi/Kerbau/Kambing/Babi,dll)"],
}
```

```python
# kumpulkan semua kolom yang perlu dibersihkan
all_cols = set(sum(P_MAP.values(), [])) | set(sum(KB_MAP.values(), [])) | set(sum(KTB_MAP.values(), []))
```

```python
# --- 4) util untuk cek "ada nilai > 0 di salah satu kolom kandidat" ---
def any_pos(df, cols, base_index):
    cols = [c for c in cols if c in df.columns]
    if not cols:
        return pd.Series(False, index=base_index)
    return df[cols].fillna(0).gt(0).any(axis=1).reindex(base_index, fill_value=False)
```

```python
# --- 5) hitung skor ---
p   = p_num.set_index(KEYS)
kb  = kb_num.set_index(KEYS)
ktb = ktb_num.set_index(KEYS)
base_index = p.index.union(kb.index).union(ktb.index)

scores = {}
```

```python
# BERAS: 100 kalau produksi / tidak beli / beli "Beras Lokal (Rp)"; 60 kalau hanya beli Beras Toko
beras_100 = (any_pos(ktb, KTB_MAP["beras"], base_index) |
             any_pos(p,   P_MAP["beras"],   base_index) |
             any_pos(kb,  ["Beras_Lokal_(Sawah/Ladang_dalam_Rp)"], base_index))
beras_60  = (~beras_100) & any_pos(kb, ["Beras_Toko"], base_index)
scores["beras"] = np.select([beras_100, beras_60], [100, 60], default=np.nan)

# JAGUNG
jagung_100 = any_pos(ktb, KTB_MAP["jagung"], base_index) | any_pos(p, P_MAP["jagung"], base_index)
jagung_60  = (~jagung_100) & any_pos(kb, KB_MAP["jagung"], base_index)
scores["jagung"] = np.select([jagung_100, jagung_60], [100, 60], default=np.nan)

# SINGKONG
singkong_100 = any_pos(ktb, KTB_MAP["singkong"], base_index) | any_pos(p, P_MAP["singkong"], base_index)
singkong_60  = (~singkong_100) & any_pos(kb, KB_MAP["singkong"], base_index)
scores["singkong"] = np.select([singkong_100, singkong_60], [100, 60], default=np.nan)

# IKAN SEGAR
ikan_100 = any_pos(ktb, KTB_MAP["ikan_segar"], base_index) | any_pos(p, P_MAP["ikan_segar"], base_index)
ikan_60  = (~ikan_100) & any_pos(kb, KB_MAP["ikan_segar"], base_index)
scores["ikan_segar"] = np.select([ikan_100, ikan_60], [100, 60], default=np.nan)

# TELOR
telor_100 = any_pos(ktb, KTB_MAP["telor"], base_index) | any_pos(p, P_MAP["telor"], base_index)
telor_60  = (~telor_100) & any_pos(kb, KB_MAP["telor"], base_index)
scores["telor"] = np.select([telor_100, telor_60], [100, 60], default=np.nan)

# DAGING
daging_100 = any_pos(ktb, KTB_MAP["daging"], base_index) | any_pos(p, P_MAP["daging"], base_index)
daging_60  = (~daging_100) & any_pos(kb, KB_MAP["daging"], base_index)
scores["daging"] = np.select([daging_100, daging_60], [100, 60], default=np.nan)
```

```python
# --- 6) rakit dataframe hasil ---
asal_bahan = (pd.DataFrame({
    "ID": [i[0] for i in base_index],
    "Desa/Kelurahan": [i[1] for i in base_index],
    **scores
})
.sort_values(["Desa/Kelurahan", "ID"])
.reset_index(drop=True))
```
