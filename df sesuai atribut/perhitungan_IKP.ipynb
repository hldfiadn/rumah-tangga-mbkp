```python
import numpy as np
import pandas as pd
import re

_num_pattern = re.compile(r'^\s*[+-]?\d+(?:[.,]\d+)?\s*$')  # dukung koma/titik desimal

def sum_semicolon_numbers(cell):
    # NaN -> NaN
    if pd.isna(cell):
        return np.nan

    s = str(cell).strip()
    if s == '':
        return np.nan  # sel benar-benar kosong

    # Normalisasi delimiter & desimal
    s = s.replace('；', ';')           # fullwidth semicolon -> normal
    parts = [p.strip() for p in s.split(';')]

    total = 0.0
    found_number = False
    for p in parts:
        if p == '':         # kosong di antara ; -> 0 (skip saja)
            continue
        p2 = p.replace(',', '.')  # dukung "0,5"
        if _num_pattern.match(p2):
            total += float(p2)
            found_number = True
        else:
            # token tidak numerik -> abaikan (anggap 0)
            # kalau mau anggap invalid -> set found_number=False dan break lalu return np.nan
            pass

    return total if found_number else np.nan
```

```python
df['Luas Lahan Keseluruhan'] = df['Luas Lahan'].apply(sum_semicolon_numbers)
df['Produksi Keseluruhan']   = df['Produksi'].apply(sum_semicolon_numbers)
```

```python
# Hapus baris dengan Luas Lahan Keseluruhan = 0
df = df[df['Luas Lahan Keseluruhan'] != 0]
```

```python
def gini(array):
    array = np.array(array, dtype=np.float64)
    if np.amin(array) < 0:
        array -= np.amin(array)
    array += 1e-9  # untuk menghindari pembagian nol
    array = np.sort(array)
    n = len(array)
    index = np.arange(1, n + 1)
    return ((np.sum((2 * index - n - 1) * array)) / (n * np.sum(array)))
```

```python
# Hitung rasio gini untuk tiap desa
rasio_gini = (
    df.groupby('Desa/Kelurahan')['Luas Lahan Keseluruhan']
    .apply(lambda x: gini(x.dropna()))
    .reset_index(name='Rasio Gini')
)
```

```python
# Fungsi scoring berdasarkan aturan di gambar
def skor_gini(g):
    if g > 0.5:
        return 0
    elif g > 0.4:
        return 1
    else:
        return 4

rasio_gini['Skor'] = rasio_gini['Rasio Gini'].apply(skor_gini)
```

```python
IKP = rasio_gini[['Desa/Kelurahan']].copy()
IKP = IKP.merge(
    rasio_gini[['Desa/Kelurahan','Skor']].rename(columns={'Skor':'I.1.1.1'}),
    on='Desa/Kelurahan', how='left'
)
```

```python
# PERHITUNGAN IKP

def sum_semicolon_numbers(cell):
    if pd.isna(cell):
        return 0.0  # kosong dianggap 0
    parts = str(cell).split(';')
    total = 0.0
    for p in parts:
        p = p.strip()
        if p == '':
            continue
        try:
            total += float(p.replace(',', '.'))
        except:
            pass
    return total

# Kolom-kolom biaya
cols_biaya = [
    'Pupuk',
    'Pestisida/Herbisida/Fungisida',
    'Benih',
    'Upah Buruh',
    'Sewa Alat dan Mesin Pertanian'
]

# Konversi semua kolom biaya ke total float
for col in cols_biaya:
    df[col] = df[col].apply(sum_semicolon_numbers)

# Hitung Ongkos Produksi
df['Ongkos Produksi'] = df[cols_biaya].sum(axis=1)
```

### II.1.3.1 Melakukan pemanfaatan limbah pertanian

```python
def posisi_pertanian(s):
    # rapikan delimiter ; dan spasi
    s = str(s).strip()
    s = re.sub(r'\s*;\s*', ';', s)
    s = re.sub(r';+$', '', s)

    # pecah jadi list
    parts = [p.strip() for p in s.split(';') if p.strip()]

    # cari posisi target
    try:
        idx = parts.index("Pertanian dan Perkebunan")
        return idx + 1   # posisi 1-based
    except ValueError:
        return 0         # kalau tidak ada target

# contoh pemakaian
data_limbah['posisi'] = data_limbah['Sumber Limbah'].apply(posisi_pertanian)
```

```python
import numpy as np

def ambil_elemen_ke(s, pos):
    parts = [p.strip() for p in str(s).split(';') if p.strip()]
    if pos > 0 and pos <= len(parts):
        return parts[pos-1]   # 1-based index
    return np.nan

# buat kolom posisi
data_limbah['posisi'] = data_limbah['Sumber Limbah'].apply(posisi_pertanian)

# ambil hanya elemen sesuai posisi
data_limbah['Sumber Limbah (posisi)'] = data_limbah.apply(
    lambda row: ambil_elemen_ke(row['Sumber Limbah'], row['posisi']), axis=1
)

data_limbah['Pengolahan (posisi)'] = data_limbah.apply(
    lambda row: ambil_elemen_ke(row['Pengolahan'], row['posisi']), axis=1
)
```
