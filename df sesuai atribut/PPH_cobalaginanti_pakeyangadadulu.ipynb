```python
kelompok_cols = {
    "padi-padian": [
        "kalori_200g_Biskuit_Roti_Kue_Kemasan",
        "kalori_kg_Beras_Lokal_Sawah_Ladang",
        "kalori_kg_Beras_Toko",
        "kalori_kg_Jagung_Hibrida",
        "kalori_kg_Jagung_Lokal",
        "kalori_kg_Jewawut_Weteng",
    ],
    "pangan hewani": [
        "kalori_kg_Ayam",
        "kalori_kg_Daging_Sapi_Kerbau_Kambing_Babi,dll",
        "kalori_kg_Ikan_dan_Boga_Laut_Kering",
        "kalori_kg_Ikan_dan_Boga_Laut_Segar",
        "kalori_kg_Susu_Bubuk_UHT",
        "kalori_kg_Telor",
    ],
    "sayur dan buah": [
        "kalori_butir_Kelapa",
        "kalori_kg_Bayam",
        "kalori_kg_Daun_Ubi",
        "kalori_kg_Jeruk",
        "kalori_kg_Kangkung",
        "kalori_kg_Kubis",
        "kalori_kg_Labu",
        "kalori_kg_Mangga",
        "kalori_kg_Pepaya",
        "kalori_kg_Pisang",
        "kalori_kg_Salak",
        "kalori_kg_Sawi",
        "kalori_kg_Timun",
        "kalori_kg_Tomat",
        "kalori_kg_Wortel",
    ],
    "umbi-umbian": [
        "kalori_kg_Singkong",
        "kalori_kg_Umbi-umbian_lain",
    ],
    "kacang-kacangan": [
        "kalori_kg_Kacang-kacangan",
        "kalori_kg_Tahu_dan_Tempe",
    ],
    "buah dan biji berminyak": [
        "kalori_ons_Kemiri",
    ],
    "minyak dan lemak": [
        "kalori_kg_Minyak_Goreng_Kemasan_Toko",
        "kalori_kg_Minyak_Kelapa",
    ],
    "gula": [
        "kalori_kg_Gula_Pasir",
        "kalori_kg_Kental_Manis",
    ],
    "lainnya (bumbu)": [
        "kalori_400g_MSG",
        "kalori_botol_Kecap",
        "kalori_bungkus_Bumbu_Kemasan",
        "kalori_kg_Bawang_Merah",
        "kalori_kg_Bawang_Putih",
        "kalori_kg_Cabai_Besar",
        "kalori_kg_Cabai_Kecil",
        "kalori_ons_Lada",
    ],
}
```

```python
def ringkas_kalori_per_kelompok(df_kal: pd.DataFrame) -> pd.DataFrame:
    id_cols = ["ID", "Desa/Kelurahan"]
    out = df_kal[id_cols].copy()

    # total per kelompok (hanya kolom yang ada)
    for nama_kelompok, cols in kelompok_cols.items():
        cols_ada = [c for c in cols if c in df_kal.columns]
        if cols_ada:
            out[f"total_kalori_{nama_kelompok}"] = df_kal[cols_ada].fillna(0).sum(axis=1)
        else:
            out[f"total_kalori_{nama_kelompok}"] = 0.0

    # Total_Kalori = jumlah dari 9 kelompok (BUKAN semua 'kalori_')
    kol_kelompok = [f"total_kalori_{k}" for k in kelompok_cols.keys()]
    out["Total_Kalori"] = out[kol_kelompok].sum(axis=1, min_count=1)

    # proporsi per kelompok (% dari total)
    for nama_kelompok in kelompok_cols.keys():
        tot_k = out[f"total_kalori_{nama_kelompok}"]
        out[f"persen_{nama_kelompok}"] = np.where(
            out["Total_Kalori"] > 0, (tot_k / out["Total_Kalori"]) * 100, 0.0
        )

    # rapikan kolom
    kol_kelompok_sorted = sorted(kol_kelompok)
    kol_persen   = sorted([f"persen_{k}" for k in kelompok_cols.keys()])
    out = out[["ID", "Desa/Kelurahan", "Total_Kalori"] + kol_kelompok_sorted + kol_persen].round(2)
    return out
```

```python
ringkasan_kelompok = ringkas_kalori_per_kelompok(df_kal)
```

```python
# daftar kategori mengikuti keys di kelompok_cols
kategori = list(kelompok_cols.keys())

# proporsi IDEAL (%)
ideal = {
    "padi-padian":             50,
    "umbi-umbian":              6,
    "pangan hewani":           12,
    "minyak dan lemak":        10,
    "buah dan biji berminyak":  3,
    "kacang-kacangan":          5,
    "gula":                     5,
    "sayur dan buah":           6,
    "lainnya (bumbu)":          3,
}

# helper: nama kolom persen & total-kalori di ringkasan
persen_col = {k: f"persen_{k}" for k in kategori}
total_col  = {k: f"total_kalori_{k}" for k in kategori}
```

```python
def skor_pph(x: float) -> int:
    if x < 40:   return 0
    if x <= 60:  return 1
    if x <= 75:  return 2
    if x <= 90:  return 3
    return 4

def hitung_pph(df_ringkas: pd.DataFrame) -> pd.DataFrame:
    df = df_ringkas.copy()

    # pastikan persen_* ada; kalau tidak, hitung dari total_kalori_*
    missing_persen = [c for c in persen_col.values() if c not in df.columns]
    if missing_persen:
        # pastikan total_kalori_* tersedia (kalau tidak ada, isi 0)
        for k in kategori:
            if total_col[k] not in df.columns:
                df[total_col[k]] = 0.0
        total9 = df[[total_col[k] for k in kategori]].sum(axis=1, min_count=1)
        for k in kategori:
            df[persen_col[k]] = np.where(
                total9 > 0,
                (df[total_col[k]] / total9) * 100.0,
                0.0
            )

    # deviasi per kategori
    for k in kategori:
        df[f"deviasi_{k}"] = (df[persen_col[k]] - ideal[k]).abs()

    # total deviasi, PPH, skor
    dev_cols = [f"deviasi_{k}" for k in kategori]
    persen_cols = [persen_col[k] for k in kategori]

    df["total_deviasi_pph"] = df[dev_cols].sum(axis=1)
    df["PPH"] = (100 - df["total_deviasi_pph"]).clip(lower=0, upper=100)
    df["Skor_PPH"] = df["PPH"].apply(skor_pph).astype("int32")

    # cek jumlah persen (harusnya ≈ 100)
    df["cek_jumlah_persen"] = df[persen_cols].sum(axis=1)

    # pembulatan
    round_cols = persen_cols + dev_cols + ["total_deviasi_pph", "PPH", "cek_jumlah_persen"]
    df[round_cols] = df[round_cols].round(2)

    # susun output minimal (tanpa kolom kalori)
    out_cols = (
        ["ID", "Desa/Kelurahan"] +
        sorted(persen_cols) +
        sorted(dev_cols) +
        ["total_deviasi_pph", "PPH", "Skor_PPH", "cek_jumlah_persen"]
    )
    out_cols = [c for c in out_cols if c in df.columns]
    return df[out_cols]
```

```python
def pph_desa_from_totals(ringkasan_kelompok: pd.DataFrame) -> pd.DataFrame:
    # kolom total_kalori_* untuk 9 kategori
    kol_tot = [f"total_kalori_{k}" for k in kategori]
    # pastikan ada (kalau belum, jalankan ringkas_kalori_per_kelompok dulu)
    missing = [c for c in kol_tot if c not in ringkasan_kelompok.columns]
    if missing:
        raise ValueError(f"Kolom hilang di ringkasan: {missing}")

    # 1) agregasi total energi per-kelompok ke level desa
    agg = (ringkasan_kelompok
           .groupby("Desa/Kelurahan", as_index=False)[kol_tot]
           .sum())

    # 2) total energi desa (hanya dari 9 kelompok)
    agg["Total_Kalori_Desa"] = agg[kol_tot].sum(axis=1, min_count=1)

    # 3) persen per-kelompok di desa
    for k in kategori:
        agg[f"persen_{k}"] = np.where(
            agg["Total_Kalori_Desa"] > 0,
            (agg[f"total_kalori_{k}"] / agg["Total_Kalori_Desa"]) * 100.0,
            0.0
        )

    # 4) deviasi & PPH desa
    for k in kategori:
        agg[f"deviasi_{k}"] = (agg[f"persen_{k}"] - ideal[k]).abs()

    dev_cols = [f"deviasi_{k}" for k in kategori]
    persen_cols = [f"persen_{k}" for k in kategori]

    agg["total_deviasi_pph"] = agg[dev_cols].sum(axis=1)
    agg["PPH_Desa"] = (100 - agg["total_deviasi_pph"]).clip(0, 100)
    agg["Skor_PPH_Desa"] = agg["PPH_Desa"].apply(skor_pph).astype("int32")
    agg["cek_jumlah_persen"] = agg[persen_cols].sum(axis=1)

    # 5) keluaran ringkas per desa (kalau mau tampilkan persen per kelompok, tambahkan persen_cols)
    out_cols = ["Desa/Kelurahan", "PPH_Desa", "Skor_PPH_Desa", "cek_jumlah_persen"]
    return agg[out_cols].round(2)
```

```python
pph_per_desa = pph_desa_from_totals(ringkasan_kelompok)
pph_per_desa
```

```python
# --- 1) kalau belum ada:
# pph_per_id = hitung_pph(ringkasan_kelompok)   # hasil: per ID ada kolom 'PPH' & 'Skor_PPH'
# pph_per_desa = pph_desa_from_totals(ringkasan_kelompok)  # hasil: 'Desa/Kelurahan','PPH_Desa','Skor_PPH_Desa'

# --- 2) (a) Rata-rata PPH & Skor_PPH per desa dari individu (ID)
pph_mean_from_id = (
    pph_per_id.groupby("Desa/Kelurahan", as_index=False)
              .agg(PPH_mean_from_ID=("PPH", "mean"),
                   Skor_mean_from_ID=("Skor_PPH", "mean"),
                   n_ID=("ID", "count"))
)

# --- 3) Gabungkan dengan hasil desa langsung (cara C)
cmp = pph_per_desa.merge(pph_mean_from_id, on="Desa/Kelurahan", how="outer")

# --- 4) Turunkan metrik pembanding
# skor dari PPH_mean (supaya apple-to-apple level skor)
cmp["Skor_from_mean_PPH"] = cmp["PPH_mean_from_ID"].apply(skor_pph).astype("Int64")

# delta (positif artinya PPH_desa > rata-rata PPH individu)
cmp["delta_PPH"]  = (cmp["PPH_Desa"] - cmp["PPH_mean_from_ID"]).round(2)
cmp["delta_Skor"] = (cmp["Skor_PPH_Desa"] - cmp["Skor_from_mean_PPH"]).astype("Int64")

# --- 5) (opsional) rangkum & cek korelasi
summary = {
    "korelasi_PPH_vs_mean": cmp[["PPH_Desa","PPH_mean_from_ID"]].corr().iloc[0,1],
    "MAE_delta_PPH": cmp["delta_PPH"].abs().mean(),
    "max_abs_delta_PPH": cmp["delta_PPH"].abs().max(),
    "desa_dengan_delta_0": int((cmp["delta_PPH"].abs() < 1e-9).sum()),
}
print(summary)

# --- 6) lihat desa yang paling beda antara cara C vs cara (a)
cmp_topdiff = cmp.sort_values("delta_PPH", key=lambda s: s.abs(), ascending=False)
```

```python
# 1) PPH desa dengan breakdown per kategori (persen & deviasi)
def pph_desa_with_breakdown(ringkasan_kelompok: pd.DataFrame) -> pd.DataFrame:
    kat = list(kategori)
    kol_tot = [f"total_kalori_{k}" for k in kat]

    # agregasi total energi per-kelompok di level desa
    agg = (ringkasan_kelompok
           .groupby("Desa/Kelurahan", as_index=False)[kol_tot]
           .sum())

    # total energi desa (9 kelompok)
    agg["Total_Kalori_Desa"] = agg[kol_tot].sum(axis=1, min_count=1)

    # persen & deviasi per kategori
    persen_cols = []
    dev_cols = []
    for k in kat:
        pcol = f"persen_{k}"
        dcol = f"deviasi_{k}"
        persen_cols.append(pcol)
        dev_cols.append(dcol)

        agg[pcol] = np.where(
            agg["Total_Kalori_Desa"] > 0,
            (agg[f"total_kalori_{k}"] / agg["Total_Kalori_Desa"]) * 100.0,
            0.0,
        )
        agg[dcol] = (agg[pcol] - ideal[k]).abs()

    # PPH desa
    agg["total_deviasi_pph"] = agg[dev_cols].sum(axis=1)
    agg["PPH_Desa"] = (100 - agg["total_deviasi_pph"]).clip(0, 100)
    # skor
    def _skor(x):
        if x < 40: return 0
        if x <= 60: return 1
        if x <= 75: return 2
        if x <= 90: return 3
        return 4
    agg["Skor_PPH_Desa"] = agg["PPH_Desa"].apply(_skor).astype("int32")
    agg["cek_jumlah_persen"] = agg[ [f"persen_{k}" for k in kat] ].sum(axis=1)

    # pilih kolom output (tampilkan persen & deviasi per kategori)
    out_cols = (["Desa/Kelurahan", "PPH_Desa", "Skor_PPH_Desa", "cek_jumlah_persen"]
                + [f"persen_{k}" for k in kat]
                + [f"deviasi_{k}" for k in kat])
    return agg[out_cols].round(2)

pph_per_desa_break = pph_desa_with_breakdown(ringkasan_kelompok)
```

```python
# 2) Rata-rata persen per-ID per desa (untuk pembanding)
mean_persen_from_id = (
    ringkasan_kelompok
      .groupby("Desa/Kelurahan", as_index=False)
      [[persen_col[k] for k in kategori]].mean()
)

# hitung deviasi dari mean persen vs ideal (di level desa)
for k in kategori:
    mean_persen_from_id[f"deviasi_meanID_{k}"] = (
        mean_persen_from_id[persen_col[k]] - ideal[k]
    ).abs()

# rapikan nama kolom mean-ID agar mudah dibedakan
rename_map = {}
for k in kategori:
    rename_map[persen_col[k]] = f"persen_meanID_{k}"
mean_persen_from_id = mean_persen_from_id.rename(columns=rename_map)

# 3) Gabungkan dengan versi desa langsung + buat delta per kategori
cmp = pph_per_desa_break.merge(mean_persen_from_id, on="Desa/Kelurahan", how="left")

for k in kategori:
    # selisih komposisi
    cmp[f"delta_persen_{k}"] = cmp[f"persen_{k}"] - cmp[f"persen_meanID_{k}"]
    # selisih “kedekatan ke ideal”: positif -> rata-rata individu lebih jauh dari ideal dibanding desa langsung
    cmp[f"delta_deviasi_{k}"] = (
        cmp[f"deviasi_meanID_{k}"] - cmp[f"deviasi_{k}"]
    )

# contoh kolom yang sering dilihat
cols_show = (["Desa/Kelurahan", "PPH_Desa", "Skor_PPH_Desa"]
             + [f"persen_{k}" for k in kategori]
             + [f"deviasi_{k}" for k in kategori]
             + [f"persen_meanID_{k}" for k in kategori]
             + [f"deviasi_meanID_{k}" for k in kategori]
             + [f"delta_persen_{k}" for k in kategori]
             + [f"delta_deviasi_{k}" for k in kategori])

cmp_view = cmp[cols_show].round(2)
```
